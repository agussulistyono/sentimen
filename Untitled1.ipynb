```python
import pandas as pd
import numpy as np
import nltk
import string
import re
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from googletrans import Translator
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def load_data():
    data = pd.read_csv('vaksincovid-19062021.csv')
    return data
```

```python
tweet_df = load_data()
tweet_df
```

```python
df = pd.DataFrame(tweet_df[['Id','Tweet']])
df
```

```python
#mengubah menjadi huruf kecil(lowe case)
df['Text'] = df['Tweet'].str.lower()
df.head()
```

```python
#data cleansing(1.menghapus user)
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt
df['remove_user'] = np.vectorize(remove_pattern)(df['Text'], "@[\w]*")
df.head(5)
```

```python
df.to_csv("remove_user.csv")
```

```python
# Data Cleansing(2. penghapusan  link, angka : 2 dll)
def remove(tweet):
    #remove angka
    tweet = re.sub('[0-9]+', '', tweet)
    
    # remove stock market tickers like $GE
    tweet = re.sub(r'\$\w*', '', tweet)
 
    # remove old style retweet text "RT"
    tweet = re.sub(r'^rt[\s]+', '', tweet)
    
    # remove hyperlinks
    tweet = re.sub(r'https?:\/\/.*[\r\n]*', '', tweet)
    
    # remove hashtags
    # only removing the hash # sign from the word
    tweet = re.sub(r'#', '', tweet)
    
    #remove coma
    tweet = re.sub(r',','',tweet)
    return tweet
df['Cleansing2'] = df['remove_user'].apply(lambda x: remove(x))
```

```python
df
```

```python
df.to_csv("hasilcleansing2.csv")
```

```python
#Data cleansing 3( membersihakan karakter yang tersisa)
def remove_punct(text):
    text = re.sub(r'[^a-zA-Z0-9]', ' ', str(text))
    text = re.sub('[0-9]+', '', text)
    text = re.sub(r'\b\w{1,2}\b', '', text) #menghilangkan 2kata
    text = re.sub(r'\s\s+', ' ', text)
    return text

df['Cleansing3'] = df['Cleansing2'].apply(lambda x: remove_punct(x))
df.head(5)
```

```python
df.drop_duplicates(subset =['Cleansing3'])
df
```

```python
#tokenization
def tokenization(text):
    text= re.split('\W+', text)
    return text
df['Tokenization']= df['Cleansing3'].apply(tokenization)

df.head()
```

```python
df.to_csv("tokenization20jun.csv")
```

```python
#Normalisasi
normalized_word = pd.read_csv("tokenization20jun.csv", encoding='latin1')

normalized_word_dict = {}

for index, row in normalized_word.iterrows():
    if row[0] not in normalized_word_dict:
        normalized_word_dict[row[0]] = row[1]

def normalized_term(document):
    return [normalized_word_dict[term] if term in normalized_word_dict else term for term in document]

df['Normalisasi'] = df['Tokenization'].apply(normalized_term)
df['Normalisasi'].head(10)
```

```python
df
```

```python
df.to_csv("Normalisasi20jun.csv")
```

```python
nltk.download('stopwords')
```

```python
#stop removal
stopword = nltk.corpus.stopwords.words('indonesian')

def stopwords(text):
    text= [word for word in text if word not in stopword]
    return text
df['Stop_Removal']= df['Normalisasi'].apply(stopwords)
df
```

```python
df.to_csv("stopRemoval20jun.csv")
```

```python
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# stemmed
def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in df['Stop_Removal']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '
            
print(len(term_dict))

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":" ,term_dict[term])
    
print(term_dict)

# apply stemmed ke dataframe
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

df['Tweet_stemmed'] = df['Stop_Removal'].apply(get_stemmed_term)
print(df['Tweet_stemmed'])
```
